```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels
```

```python
# Step 2: Preprocess the data (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 3: Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Step 4: Visualize the reduced features (2D scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k',
s=100)
plt.title('PCA of Iris Dataset (2 Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target Class')
plt.show()
# Optional: Explained Variance Ratio
print("Explained Variance Ratio of Each Component:")
print(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
# 3D visualization using matplotlib
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='viridis',
edgecolor='k', s=100)
ax.set_title('PCA of Iris Dataset (3 Components)')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.show()
```

PCA on tiitanic data

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("/content/titanic .csv")
```

```python
df = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex']]
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)


```

```python
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title('PCA of Titanic Dataset (2 Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Survival (0 = No, 1 = Yes)')
plt.show()
```

```python
print("Explained Variance Ratio of Each Component:")
print(pca.explained_variance_ratio_)

```

```python
# Step 5: Apply PCA to reduce to 3 dimensions
pca = PCA(n_components=3)
X_pca_3D = pca.fit_transform(X_scaled)

```

```python
# Step 6: 3D Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca_3D[:, 0], X_pca_3D[:, 1], X_pca_3D[:, 2], c=y, cmap='viridis', edgecolor='k', s=100)
ax.set_title('PCA of Titanic Dataset (3 Components)')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component3')
plt.show()
```
